List comprehention offers a shorter syntax whenyou want to create a new list based on the values of an existing list

```python
list1 = range(10,50)

even = []

for i in list1:

    if i%2 == 0:
        even.append(i)

print(even)
```

**List Coprehansion - Shorter Syntax**

```python
list1 = range(10,50)





even = [i for i in list1 if i%2 == 0]

print(even)
```

**Global Scope and Local Scope **

```python
# A variable is only available from inside the region it is created. This is called scope
```

```python
#Local Scope
#A local created inside a function belongs to the local scope of the function ,and cn only beused inside that function.
```

```python
def myfunc():
    y = 300
    print(y)

myfunc()
```

```python
def myfunc():
    y= 300
    print(y)

myfunc()
print(y)
```

```python
def myfunc():
    y = 300
    def myinnerfunc():
        print(y)
    myinnerfunc()

myfunc()
```

```python

#Global Scope
#variable created in the main body of the python code is a global variable and belongs to the global scope.

#Global variables are available from within any scope, global and local#A var
```

```python
x = 300

def myfunc():
    print(x)

myfunc()

print(x)
```

**Global Keyword**

* if you need to create a global variable ,but are stuck in the local scope,you can use the global key word

*The global Keyword makes the variable global

```python
def myfunc():

    x = 300

myfunc()

print(x)
```

*Also ,use the global keyword if you want to make a change to a gobal variable inside

```python
def myfunc():
    global x
    x = 300

myfunc()

print(x)
```
